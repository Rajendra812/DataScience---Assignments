Outlier Detection

```python
data.hist()
```

```python
data.skew()
```

```python
data_NA = data.dropna()
```

```python
data_NA.info()
```

```python
Q1 = data_NA.quantile(0.25)
Q3 = data_NA.quantile(0.75)
IQR = Q3 - Q1

# Consider values outside 1.5 * IQR as outliers
outliers = ((data_NA < (Q1 - 1.5 * IQR)) | (data_NA > (Q3 + 1.5 * IQR))).any(axis=1)
print("Number of outliers detected:", outliers.sum())
```

```python
# Removing outliers
data_cleaned = data_NA[~outliers]
```

```python
#  Statistical summary of the cleaned dataset
summary = data_cleaned.describe()
print(summary)
```

```python
# Highlighting interesting findings
print("Interesting findings from the summary:")
print("Mean values:")
print(summary.loc['mean'])
print("Standard deviation values:")
print(summary.loc['std'])
```

```python
# Histograms for numerical variables
data_cleaned.hist(figsize=(14, 10), bins=20)
plt.tight_layout()
plt.show()
```

```python
# Boxplots for numerical variables
plt.figure(figsize=(14, 10))
sns.boxplot(data=data_cleaned)
plt.xticks(rotation=90)
plt.show()
```

```python
# Pairplot to explore relationships between pairs of variables
sns.pairplot(data_cleaned)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data_cleaned.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```

```python
# Correlation matrix
correlation_matrix = data_cleaned.corr()
print("Correlation matrix:")
print(correlation_matrix)
```

```python
# Highlight strong correlations (greater than 0.7 or less than -0.7)
strong_correlations = correlation_matrix[(correlation_matrix > 0.7) | (correlation_matrix < -0.7)]
print("Strong correlations:")
print(strong_correlations)
```
